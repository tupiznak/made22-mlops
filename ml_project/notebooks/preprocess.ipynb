```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("../data/raw/heart_cleveland_upload.csv")
X_input = df.drop(columns=["condition"])
y = df["condition"]
X_input
```

```python
CAT_FEATURES_ONE_HOT = ["sex", "cp", "restecg", "thal"]
CAT_FEATURES_LABEL = ["fbs", "exang", "slope", "ca"]
```

```python
categorial_features = CAT_FEATURES_ONE_HOT
encoder = OneHotEncoder()
encoder.fit(X_input[categorial_features])
encoder.categories_
X_transform = pd.concat(
    (
        X_input.drop(categorial_features, axis=1),
        pd.DataFrame(
            data=encoder.transform(X_input[categorial_features]).toarray(),
            columns=encoder.get_feature_names_out(),
        ),
    ),
    axis=1,
)
X_transform
```
